### 1. Scatterplot

```python
# Load libraries
import squarify
import seaborn as sb
```

```python
# Plot scatterplot visualization
plt.rcParams["figure.figsize"] = (15,10)
plt.scatter(crime_df['population'], crime_df['aggravated_assault'], color ="red", alpha=.8) 
plt.title('Aggravated Assault Crime Rate by State Population',loc='center', fontsize=16, color='k')
plt.xlabel('State Population', fontsize=14, color='k')
plt.ylabel('Aggravated Assaults', fontsize=14, color='k')
plt.show()
```

### 2. Bubble Chart

```python
# Create normalized population
pop_normal = crime_df['population']/crime_df['population'].max()
```

```python
# Plot bubble chart visualization
plt.scatter(crime_df['aggravated_assault'], crime_df['murder'], s=pop_normal*2000, color = 'red', alpha=0.8)
plt.box(False)
plt.title('Murder and Aggravated Assault Crime Rates by State Population', loc='center', fontsize=16, color='k')
plt.xlabel('Aggravated Assaults', fontsize=14, color='k')
plt.ylabel('Murders', fontsize=14, color='k')
for i, txt in enumerate(crime_df['state']):
    plt.annotate(txt, (crime_df['aggravated_assault'][i], crime_df['murder'][i]), color='k', fontsize=12)
```

### 3. Density Plot

```python
# Build density plot
ax = sb.kdeplot(np.array(crime_df['aggravated_assault']), color="red")
ax.set(xlabel='Aggravated Assaults')
sb.despine()
ax.set(title='United States Aggravated Assault Crime Rates')
```
