```python
# Load libraries
import squarify
import seaborn as sb
```

```python
# Plot tree map visualization
plt.figure(figsize=(15,6))
squarify.plot(sizes=trim_df['expenditure'], label=trim_df['category'], alpha=.8, 
              pad = True, color = sb.color_palette("Spectral"),text_kwargs={'fontsize':9},)
plt.axis('off')
plt.title('2008 Expenditures by Category',loc='center', fontsize=18, color='k')
plt.show()
```

### 2. Area Chart

```python
# Group expenditure by year on original dataframe
group_df = spend_df.groupby('year').sum()

# Reset index
group_df = group_df.reset_index()

# Verify results
group_df
```

```python
# Plot area chart visualization
plt.figure(figsize=(15,6))
plt.fill_between(group_df['year'], group_df['expenditure'], color="green", alpha=0.4)
plt.plot(group_df['year'], group_df['expenditure'], color="green", alpha=0.6, linewidth=3)
plt.box(False)
plt.title('Expenditures by Year', loc='center', fontsize=24, color='k')
plt.xlabel('Year', fontsize=16, color='k')
plt.ylabel('Expenditure', fontsize=16, color='k')
plt.tick_params(axis='x', colors='dimgray')
plt.tick_params(axis='y', colors='dimgray')
plt.show()
```

### 3. Stacked Area Chart

```python
# Create list of categories
for c in spend_df['category'].unique():
    print(c)
```

```python
# Build individual dataframes for every category
df_food = spend_df.loc[spend_df['category']=='Food']
df_alcohol = spend_df.loc[spend_df['category']=='Alcoholic Beverages']
df_housing = spend_df.loc[spend_df['category']=='Housing']
df_apparel = spend_df.loc[spend_df['category']=='Apparel']
df_transportation = spend_df.loc[spend_df['category']=='Transportation']
df_healthcare = spend_df.loc[spend_df['category']=='Healthcare']
df_entertainment = spend_df.loc[spend_df['category']=='Entertainment']
df_personalcare = spend_df.loc[spend_df['category']=='Personal Care']
df_reading = spend_df.loc[spend_df['category']=='Reading']
df_education = spend_df.loc[spend_df['category']=='Education']
df_tobacco = spend_df.loc[spend_df['category']=='Tobacco Products']
df_miscellaneous = spend_df.loc[spend_df['category']=='Miscellaneous']
df_cash = spend_df.loc[spend_df['category']=='Cash Contributions']
df_personalinsurance = spend_df.loc[spend_df['category']=='Personal Insurance']
```

```python
# Build stacked area chart on the multiple dataframes
plt.stackplot(df_food['year'], df_food['expenditure'], df_alcohol['expenditure'], df_housing['expenditure'], 
              df_apparel['expenditure'], df_transportation['expenditure'], df_healthcare['expenditure'], 
              df_entertainment['expenditure'], df_personalcare['expenditure'], df_education['expenditure'], 
              df_tobacco['expenditure'], df_miscellaneous['expenditure'], df_cash['expenditure'], 
              df_personalinsurance['expenditure'], labels=['Food', 'Alcoholic Beverages', 'Housing', 'Apparel', 
                                                           'Transporation', 'Healthcare', 'Entertainment', 'Personal Care',
                                                          'Reading', 'Education', 'Tobacco Products', 'Miscellaneous', 
                                                          'Cash Contributions', 'Personal Insurance'])
plt.legend(loc='upper left', fontsize=12)
plt.title('Expenditures by Year and Category', loc='center', fontsize=24, color='k')
plt.xlabel('Year', fontsize=12, color='k')
plt.ylabel('Expenditure', fontsize=12, color='k')
plt.tick_params(axis='x', colors='dimgray')
plt.tick_params(axis='y', colors = 'dimgray')
plt.rcParams["figure.figsize"] = (18,10)
```

### 4. Step Chart

```python
# Plot step chart visualization
plt.step(group_df['year'], group_df['expenditure'], color="green", alpha=0.6, linewidth=3)
plt.box(False)
plt.title('Expenditures by Year', loc='center', fontsize=24, color='k')
plt.xlabel('Year', fontsize=16, color='k')
plt.ylabel('Expenditure', fontsize=16, color='k')
plt.tick_params(axis='x', colors='dimgray')
plt.tick_params(axis='y', colors='dimgray')
plt.show()
```
